```python
sns.barplot(x='Rating',y='Title',data=top_rate,hue='Director',dodge=True)
plt.legend(bbox_to_anchor=(1.1,1),loc=2)
"""
hue='Director'
Bars are colored based on the Director
Different directors get different colors

This separates bars side-by-side when there are multiple hue categories.
Without dodge:
Bars overlap
With dodge=True:
Bars appear beside each other

bbox_to_anchor=(1.05,1)
Moves legend outside the graph.
1.05 → slightly right of plot
1 → top aligned, 0-> bottom aligned
So legend appears outside on right side.

Think of the legend as a sticker.
bbox_to_anchor gives the wall position.
loc decides:
which corner of the sticker touches the wall.

loc value	Meaning
0	         best
1	       upper right
2	       upper left
3	       lower left
4	       lower right
8	       lower center
9	       upper center
"""
```

<span style='font-size:20px'>**Classify Movies based on ratings [excellent,good,average]**</span>

```python
data.columns
```

```python
def rating(rating):
    if rating>=7.0:
        return "Excellent"
    elif rating>=6.0:
        return "Good"
    else:
        return "Average"
```

```python
data['rating_cat']=data['Rating'].apply(rating)
```

<span style='font-size:20px'>**Find the unique values from genre**</span>

```python
data.columns
```

```python
list1=[]
for value in data['Genre']:
    list1.append(value.split(','))
list1
```

```python
#coverting 2D list to 1d List
one_d=[]
for item in list1:
    for item2 in item:
        one_d.append(item2)
```

```python
#repeatation of genre
one_d
```

```python
#we are creating list with unique genre not the repeative genre
uni_list=[]
for item in one_d:
    if item not in uni_list:
        uni_list.append(item)
```

```python
uni_list
```

```python
len(uni_list)
```
